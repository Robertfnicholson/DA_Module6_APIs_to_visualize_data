```python
# Import linregress and pyplot
from scipy.stats import linregress
import matplotlib.pyplot as plt
```

```python
# Chicago temps and daylight, 12 random days
temps = [18.2, 29.5, 35.9, 50.1, 58.6, 69.9, 82.1, 85.5, 73.5, 55.2, 40.3, 30.1]
daylight = [7.6, 9.2, 12.2, 13.4, 13.9, 14.2, 13.5, 11.8, 10.9, 9.9, 8.4, 7.8]
```

```python
# Plot the data with daylight as the independent variable
plt.scatter(daylight,temps)
plt.xlabel("Hours of Daylight")
plt.ylabel("Average Monthly Temperature")
plt.show()
```

```python
# Use linear regression on the data
linregress(daylight,temps)
```

```python
# Create a tuple that is the same size as the linear regression output and set equal to the linear regression function
(slope, intercept, rvalue, pvalue, stderr) = linregress(daylight,temps)
```

```python
# Create the linear regression line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
# to add our linear equation model to our scatter plot. Use list compreshension is the easiest in Python
regress_values = [(day * slope + intercept) for day in daylight]
```

```python
plt.scatter(daylight,temps)
plt.plot(daylight,regress_values)
plt.xlabel("Hours of Daylight")
plt.ylabel("Average Monthly Temperature")
plt.text(10,65,line_eq, style='italic')
plt.show()
```
